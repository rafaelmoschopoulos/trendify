```python
import pandas as pd
dataset_path = './data/sentiment140.csv'
df = pd.read_csv(dataset_path, names = ['polarity', 'id', 'date', 'query', 'user', 'text'], index_col = None,  encoding='latin-1')
```

```python
df = df.sample(n=10000)
```

```python
from textcleaner import TextCleaner
cleaner = TextCleaner()

df.loc[:, 'text'] = df['text'].apply(lambda doc: cleaner.clean(doc, False))
df.to_csv('./data/cleaned_notrans.csv', index=False, encoding="utf-8")
df
```
